```python
caracteristicas = ['user_id','title','items_count','playtime_forever','developer','genres','price','anio_lanzamiento','posted','item_id','recommend','sentiment_analysis']
filtro = df_ML[caracteristicas]
filtro
```

```python
filtro[caracteristicas] = filtro[caracteristicas].astype(str)
```

```python
for caracteristica in caracteristicas:
    filtro[caracteristica]=filtro[caracteristica].apply(lambda x: x.lower().replace(" ",""))
```

```python
filtro.shape
```

```python
def crear_sopa(dato):
    return dato['user_id']+ " "+ dato['title']+ " " + dato['items_count']+ " " + dato['playtime_forever']+ " " + dato['developer']+ " " + dato['genres']+ " "+ dato['price']+ " "+ dato['anio_lanzamiento']+ " "+ dato['posted']+ " "+ dato['item_id']+ " "+ dato['recommend']+ " "+ dato['sentiment_analysis']
```

```python
filtro['caracteristicas']=filtro.apply(crear_sopa, axis=1)
```

```python
filtro
```

```python
filtro[caracteristicas]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

modelo=TfidfVectorizer(stop_words='english')
matriz_modelo = modelo.fit_transform(filtro[caracteristicas])
```

```python
modelo
```

```python
matriz_modelo
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similitudCoseno=cosine_similarity(X=matriz_modelo, Y=matriz_modelo)
```

```python
similitudCoseno.shape
```
