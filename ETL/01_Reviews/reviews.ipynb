#### Creando la columna Sentiment

```python
from textblob import TextBlob

def analiZentimiento(texto):
    analysis = TextBlob(texto) # crear un objeto TextBlob del texto
    if analysis.sentiment.polarity < 0:
        return 0 # 'Negativo'
    elif analysis.sentiment.polarity == 0:
        return 1 # 'Neutral'
    else:
        return 2 # 'Positivo'
    
rev_reviews['sentiment_analysis'] = rev_reviews['review'].astype(str).apply(analiZentimiento)
rev_reviews['sentiment_analysis'].fillna(1,inplace=True)
        
```
