#### Desanidando la columna items.

```python
items_items = {'user_id':[], 'item_id':[], 'item_name':[], 'playtime_forever':[], 'playtime_2weeks':[]}
for index, array in enumerate(items['items']):
    for item in eval(str(array)):
        for clave, valor in item.items():
            items_items[clave].append(valor)
        items_items['user_id'].append(items.iloc[index]['user_id'])
items_items = pd.DataFrame(items_items)
items_items
```

#### Uniendo los dataframe.

```python
itemsMarge =(pd.merge(items_items, items, left_on="user_id", right_on="user_id")[["user_id","items_count","item_id","item_name","playtime_forever","playtime_2weeks"]])
itemsMarge.to_parquet(r'C:\Users\ingju\Desktop\jrPI_1\data\02Items\04_itemsMarge')
itemsMarge.head()
```
