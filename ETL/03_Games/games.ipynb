### Games.json

#### Importando librerias.

```python
import pandas as pd
import numpy as np
import sys
```

#### Leyendo el archivo y crando el DataFrame.

```python
df = []
with open(r'C:\Users\ingju\Desktop\jrPI_1\data\00Original\output_steam_games.json', encoding='utf-8-sig') as file:
    for linea in file:
        if ' "id":NaN' in linea:
            continue
        linea = linea.replace('false', 'False')
        linea = linea.replace('true', 'True')
        linea = linea.replace('NaN', 'None')
        df.append(eval(linea))
games = pd.DataFrame(df)
```

```python
games.head()
```

#### Ver si hay si hay duplicados.

```python
games[games['id'].duplicated(keep=False)]
```

#### Eliminado duplicados.

```python
games.drop_duplicates(subset='id', keep='first',inplace=True)
```

```python
games.head()
```

#### Convirtiendo las filas de la columna release_date a tipo datetime

```python
games['release_date']= pd.to_datetime(games['release_date'], format= '%Y-%m-%d', errors='coerce')
```

```python
games.head()
```

```python
games['release_date'] = games['release_date'].dt.year
```

```python
games.head()
```

#### Eliminando valores nulos de la columna release_date

```python
games['release_date'] = games['release_date'].dropna()
```

```python
games.shape
```

#### Eliminnado el resto de los valores nulos.

```python
games = games.dropna()
games.head()
```

```python
games.shape
```

#### Funcion para convertir ceros.

```python
def ceros(x):
    try:
        return float(x)
    except:
        return (0)
games['price']= games['price'].apply(ceros)
games['price']
```

```python
games.head()
```

#### Cambiando el nombre a la columna id por item_id

```python
games.rename(columns={'id':'item_id'}, inplace=True)
```

```python
games.sample(5)
```

```python
games.shape
```

#### Eliminando filas que contienen valores nulos en la columna item_id

```python
filtro = games[games.item_id.isnull()].index
games.drop(filtro, inplace=True)
games.reset_index(inplace=True, drop=True)
games.shape
```

#### Eliminando filas que contienen valores nulos en la columna genres

```python
filtro = games[games.genres.isnull()].index
games.drop(filtro, inplace=True)
games.reset_index(inplace=True, drop=True)
games.shape
```

#### Creando columna anio de lanzamiento.

```python
games['anio_lanzamiento'] = games['release_date'].apply(lambda x: str(x)[:4] if str(x)[:4].isnumeric() else 0)
games.head()
```

#### Eliminando columna release_date

```python
games.drop(['release_date'], axis=1, inplace=True)
games.shape
```

#### Observando valores faltantes.

```python
games.isna().sum()
```

#### Eliminando columna app_name.

```python
games.drop(['app_name'], axis=1, inplace=True)
```

```python
games.shape
```

#### Guardo el DataFrame en un archivo .csv.

```python
games.to_csv(r'C:\Users\ingju\Desktop\jrPI_1\data\03Games\01_games.csv', index= False, errors='replace')
```
